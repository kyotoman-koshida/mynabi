### 足立区

```python
df_adachi = pd.read_csv('../data/outer_data/tokyo-adachi.csv', encoding='shift-jis')
print(df_adachi.head())

#所在地のテキスト整形用関数
def func_place(block):
    target2 = '丁'
    target3 = '町'
    target4 = '-'
    target5 = '－'
    try:
        if target2 in block:
            place = block.split(target2)[0]
        elif target3 in block:
            place = block.split(target3)[0]
            place = place + target3
        elif target4 in block:
            place = block.split(target4)[0]
            place = place[:-1]
        elif target5 in block:
            place = block.split(target5)[0]
            place = place[:-1]
        if place[-1].isnumeric():
            place = place[:-1]
        return place
    except:
            return block

df_adachi = pd.read_csv('../data/outer_data/tokyo-adachi.csv', encoding='shift-jis')

locations = df_adachi[['市区町村','2019年']]

#以下の部分では訓練データの”区”のラベル化に備えて、〇〇区の部分を抽出する。
i = 4
adachi2019 = []
for i in range(4, len(locations), 1):
    target1 = "都"
    idx1 = locations.loc[i]['市区町村'].find(target1)
    block = locations.loc[i]['市区町村'][idx1+1:]
    address = func_place(block)
    land_price = locations.loc[i]['2019年'].replace(',','')#カンマ区切りの数字はすぐに数値に変換できない。
    try:
        land_price = int( land_price )
    except:
        land_price = nan
    address_price = [address, land_price]
    adachi2019.append(address_price)

adachi2019 = pd.DataFrame(adachi2019)
adachi2019 = adachi2019.rename(columns={0:'区町名',1:'地価'})#列名の振り直し

##「区町名」でグループ化を施した上で、一旦ふつうのリストに区町名を順番を保って保存する。
###その上でグループ化結果の区町名ごとの平均化された地価のデータと結合させる。
####まどろっこしいが、これより簡単な方法がわからない・・・
grouped = adachi2019.groupby('区町名')#ここでグループ化
blocks = list(grouped.groups)#順番を保った上で「区町名」を全て取り出す
block_aveprice_list = []
i = 0
for block in blocks:
    aveprice = grouped.mean()['地価'][i]#区町名の順番ずつに、その平均地価を取得する。
    block_aveprice_list.append([block, aveprice])
    i += 1

block_aveprice_adachi2019 = pd.DataFrame(block_aveprice_list)#ここで改めてデータフレーム型に変換する。
block_aveprice_adachi2019.rename(columns={0:'区町名',1:'平均地価'})#列名の振り直し

block_aveprice_adachi2019.to_csv('../data/handmade_data/block_aveprice_adachi2019.csv', index=False)
```
